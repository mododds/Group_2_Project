```python
#import dependencies
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import time
import requests
import io
import os
```

```python
def create_date_based_dataframes(Symbols,start, end):
    # create empty dataframe
    stocks_df = pd.DataFrame()
    # iterate over each symbol
    for i in Symbols:  

        # print the symbol which is being downloaded
        print( str(Symbols.index(i)) + str(' : ') + i, sep=',', end=',', flush=True)  

        try:
            # download the stock price 
            stock = []
            stock = yf.download(i,start=start, end=end, progress=False)

            # append the individual stock prices 
            if len(stock) == 0:
                None
            else:
                stock['TICKER']=i
                stocks_df = stocks_df.append(stock,sort=False)
                print(f'{i}: data appended')
                
        except Exception:
            None
            
    return stocks_df
```

```python
file_to_load = os.path.join("Resources", "Stocks_with_Sector.csv")
temp_df = pd.read_csv(file_to_load)
temp_df
```

```python
Symbols = temp_df['TICKER'].tolist()
Symbols
```

```python
# Time series data start and end date
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2022,3,16)
stocks_df = create_date_based_dataframes(Symbols,start, end)
stocks_df
```

```python
stocks_df['Date'] = stocks_df.index
stocks_df
```

```python
stock_file_to_save = os.path.join("Resources", "Stocks_with_time.csv")
stocks_df.to_csv(stock_file_to_save, index =False)
```
