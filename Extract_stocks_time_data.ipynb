```python
#import dependencies
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import time
import requests
import io
import os
```

```python
# function to get the time based data from yahoo finance and storing it in a Dataframe
def create_date_based_dataframes(Symbols,start, end):
    # create empty dataframe
    stocks_df = pd.DataFrame()
    # iterate over each symbol
    for i in Symbols:  

        # print the symbol which is being downloaded
        print( str(Symbols.index(i)) + str(' : ') + i, sep=',', end=',', flush=True)  

        try:
            # download the stock price 
            stock = []
            stock = yf.download(i,start=start, end=end, progress=False)

            # append the individual stock prices 
            if len(stock) == 0:
                None
            else:
                stock['TICKER']=i
                stocks_df = stocks_df.append(stock,sort=False)
                print(f'{i}: data appended')
                
        except Exception:
            None
            
    return stocks_df
```

```python
# load csv files with details about companies to explore the time series data on
file_to_load = os.path.join("Resources", "Stocks_with_Sector.csv")
temp_df = pd.read_csv(file_to_load)
temp_df
```

```python
# Convert company tickers as list to pass to yahoo finance 
Symbols = temp_df['TICKER'].tolist()
Symbols
```

```python
# Time series data start and end date
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2022,3,16)

# Create the stocks_df
stocks_df = create_date_based_dataframes(Symbols,start, end)
stocks_df
```

```python
# including date as a column in stocks_df
stocks_df['Date'] = stocks_df.index
stocks_df
```

```python
# Saving the time series stock data to csv
stock_file_to_save = os.path.join("Resources", "Stocks_with_time.csv")
stocks_df.to_csv(stock_file_to_save, index =False)
```

```python
# Created a function to extract stocks description from yahoo finance
def create_info(Symbols):
    info =[]
    # iterate over each symbol
    for i in Symbols:  
        try:
            # download stock info
            temp =[]
            temp = yf.Ticker(i).info
            # append the individual stock info
            if len(temp) == 0:
                None
            else:
                info.append(temp)
                print(f'{i}: info appended')
                
        except Exception:
            None
            
    return info
```

```python
# Called the function to create a list of dictionaries containing stocks information
info = create_info(Symbols)
info
```

```python
# Converting the list to a dataframe
info_df = pd.DataFrame()
for ticker in info:
    info_df = info_df.append(ticker,sort=False, ignore_index=True)
    
info_df
```
