```python
#import dependencies
import pandas as pd
import numpy as np
import os
```

```python
# Loading csv files
info_file_to_load = os.path.join("Resources", "Stocks_info.csv")
esg_file_to_load = os.path.join("Resources", "Stocks_with_ESG.csv")
time_file_to_load = os.path.join("Resources", "Stocks_with_time.csv")
info_df = pd.read_csv(info_file_to_load)
esg_df = pd.read_csv(esg_file_to_load)
time_df = pd.read_csv(time_file_to_load)
```

```python
# Check the size of dataframes
print(info_df.shape)
print(esg_df.shape)
print(time_df.shape)
```

```python
# time_df unique stocks
print(time_df['TICKER'].nunique())
```

```python
# remove stocks with no information from info_df
info_df.drop(info_df.index[109], inplace=True)
print(info_df.shape)
info_df
```

```python
# remove the same stock from esg_df
esg_df.drop(esg_df.index[109], inplace=True)
print(esg_df.shape)
esg_df
```

```python
# Creating a list of tickers
tickers = esg_df['TICKER'].to_list()
print(len(tickers))
```

```python
#Checking that all the Dataframes have the same number of stocks
print(time_df['TICKER'].nunique())
print(esg_df['TICKER'].nunique())
print(info_df['TICKER'].nunique())
```

```python
# Saving the ESG parameters as a list
esg_parameters = esg_df.columns.to_list()
esg_parameters
```

```python
# Creating Society dataframe
society_df = esg_df.iloc[:,0:13]
society_df
```

```python
# Creating Knowledge dataframe
knowledge_df = esg_df.iloc[:,13:23]
knowledge_df['TICKER'] = society_df['TICKER']
shiftPos = knowledge_df.pop("TICKER")
knowledge_df.insert(0, "TICKER", shiftPos)
knowledge_df
```

```python
# Creating health dataframe
health_df = esg_df.iloc[:,23:35]
health_df['TICKER'] = society_df['TICKER']
shiftPos = health_df.pop("TICKER")
health_df.insert(0, "TICKER", shiftPos)
health_df
```

```python
# Creating environment dataframe
environment_df = esg_df.iloc[:,35:47]
environment_df['TICKER'] = society_df['TICKER']
shiftPos = environment_df.pop("TICKER")
environment_df.insert(0, "TICKER", shiftPos)
environment_df
```

```python
# Creating net_impact dataframe
net_impact_df = esg_df.iloc[:,47:]
net_impact_df['TICKER'] = society_df['TICKER']
shiftPos =net_impact_df.pop("TICKER")
net_impact_df.insert(0, "TICKER", shiftPos)
net_impact_df
```

```python
# Saving csv files as an input to postgres
info_file_to_save = os.path.join("Resources/Postgres_Input", "Stocks_info.csv")
esg_file_to_save = os.path.join("Resources/Postgres_Input", "Stocks_ESG.csv")
time_file_to_save = os.path.join("Resources/Postgres_Input", "Stocks_time.csv")
society_file_to_save = os.path.join("Resources/Postgres_Input", "society_ESG.csv")
health_file_to_save = os.path.join("Resources/Postgres_Input", "health_ESG.csv")
knowledge_file_to_save = os.path.join("Resources/Postgres_Input", "knowledge_ESG.csv")
environment_file_to_save = os.path.join("Resources/Postgres_Input", "environment_ESG.csv")
net_impact_file_to_save = os.path.join("Resources/Postgres_Input", "net_impact_ESG.csv")

info_df.to_csv(info_file_to_save, index =False)
esg_df.to_csv(esg_file_to_save, index =False)
time_df.to_csv(time_file_to_save, index =False)
society_df.to_csv(society_file_to_save, index =False)
health_df.to_csv(health_file_to_save, index =False)
knowledge_df.to_csv(knowledge_file_to_save, index =False)
environment_df.to_csv(environment_file_to_save, index =False)
net_impact_df.to_csv(net_impact_file_to_save, index =False)
```
